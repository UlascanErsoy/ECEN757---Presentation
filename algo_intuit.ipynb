```python
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import numpy as np
```

```python
x = np.linspace(0, 2, 1000)
y = np.linspace(0, 2, 1000)
colors = ['red','green']
cmap = ListedColormap(colors)

X, Y = np.meshgrid(x,y)
for C in [1.5]:
    tot = np.array([X.ravel(), Y.ravel()]).T
    tot = np.array(list(map(lambda x: 1 if x[0]+x[1] <= C and x[0] <= 1 and x[1] <= 1 else 0, list(tot))))
    tot = tot.reshape(X.shape)
    plt.contourf(X,Y, tot, cmap=cmap, alpha=0.2)

plt.xlabel("$z^{(1)}$")
plt.ylabel("$z^{(2)}$")
```
